## EXERCISE 1: MARKOV CHAIN STORY GENERATION

In this exercise, a small training passage is used to build a **bigram-based Markov chain**. The model predicts the next word based only on the previous two words. This method is simple and fast, but it does not understand meaning or long-range context.

```python
import random
import re
from collections import defaultdict

# Training text
TRAINING = """
Once upon a time there was a kingdom far away in the mountains. The kingdom was ruled by a wise and kind king.
The king had a daughter who was brave and clever. One day a dragon appeared near the kingdom and frightened the villagers.
The brave princess decided to face the dragon alone. She walked through the dark forest and found the dragon sleeping near a cave.
The princess spoke gently to the dragon and discovered it was lonely and lost. She invited the dragon to live near the kingdom.
From that day on the dragon protected the kingdom and the princess and the dragon became the best of friends.
The king was proud of his daughter and the kingdom lived happily ever after.
"""

# Build Markov chain (bigrams)
words = re.findall(r'\w+', TRAINING.lower())
chain = defaultdict(list)
for i in range(len(words) - 2):
    key = (words[i], words[i + 1])
    chain[key].append(words[i + 2])

# Generate story
def generate_story(chain, start, num_words=50):
    words = re.findall(r'\w+', start.lower())
    story = list(words)
    key = (story[-2], story[-1])

    for _ in range(num_words):
        next_words = chain.get(key)
        if not next_words:
            break
        next_word = random.choice(next_words)
        story.append(next_word)
        key = (story[-2], story[-1])

    story[0] = story[0].capitalize()
    return ' '.join(story) + '.'

story = generate_story(chain, 'Once upon a time there was a kingdom', 50)

print('=' * 70)
print('MARKOV CHAIN GENERATED STORY')
print('=' * 70)
print(story)
```

### Interpretation of Exercise 1
The generated story reuses words and sentence fragments from the training text. This shows that the Markov chain can imitate the style and vocabulary of the original passage. However, because it only uses local word transitions, the story may contain repetition, abrupt jumps, or incomplete semantic flow.

## EXERCISE 2: LLM STORY GENERATION

In this exercise, **GPT-2** is used to generate a creative story. Unlike the Markov chain, GPT-2 is a pretrained neural network that has learned language patterns from massive text corpora. It can maintain broader context and produce more fluent and meaningful output.

```python
from transformers import pipeline

print('Loading GPT-2 model...')

# Create generator
generator = pipeline('text-generation', model='gpt2')

prompt = (
    'Write a creative story about why it is important to study data science today. '
    'Once upon a time, in a world drowning in data, a young student named Alex '
    'decided to learn data science. Alex discovered'
)

result = generator(
    prompt,
    max_new_tokens=300,
    num_return_sequences=1,
    temperature=0.9,
    top_p=0.95,
    do_sample=True,
    repetition_penalty=1.2,
    pad_token_id=50256
)

print('=' * 70)
print('GPT-2 GENERATED STORY')
print('=' * 70)
print(result[0]['generated_text'])
```

### Interpretation of Exercise 2
GPT-2 typically produces more fluent and natural language than the Markov chain model. It can follow the prompt more effectively and generate content that sounds more human-like. Even so, GPT-2 can still produce occasional repetition, vague statements, or off-topic continuations depending on sampling settings.

## COMPARISON: MARKOV CHAIN vs LLM

| Aspect | Markov Chain | LLM (GPT-2) |
|--------|--------------|-------------|
| **Model type** | Statistical (bigrams) | Neural network |
| **Parameters** | Very few / rule-based transitions | ~117M parameters |
| **Understanding** | No semantic understanding | Captures semantic and contextual patterns |
| **Context window** | 2 words | Much larger token context |
| **Coherence** | Limited and often repetitive | More fluent and meaningful |
| **Speed** | Very fast | Slower |
| **Training data needed** | Small sample text | Massive pretrained corpus |

### Key Insight
The Markov chain approach is useful for demonstrating the basic idea of probabilistic text generation, but it does not truly understand language. GPT-2, as a Large Language Model, is much more effective at generating coherent and context-aware text. Therefore, LLMs are far more suitable for realistic story generation and advanced NLP applications.
